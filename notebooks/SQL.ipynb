### Exercícios

#### 1.1) Quantas estações no total existem na tabela *stations*?

```python
con.execute("""

    SELECT
        COUNT(DISTINCT station_id)
    FROM
        stations

""").fetchdf()
```

#### 1.2) Quantas estações estão com o status ativo?



```python
con.execute("""

    SELECT
        status,
        COUNT(DISTINCT station_id) AS qtd
    FROM
        stations
    GROUP BY
        status
    ORDER BY
        2 DESC
""").fetchdf()
```

#### 2.1) Conte o total de viagens da base *trips*

#### 2.2) Existem **trip_id** repetidos na tabela *trips*?

```python
con.execute("""

    SELECT
        COUNT(*) AS total,
        COUNT(DISTINCT trip_id) AS distintos
    FROM
        trips

""").fetchdf()
```

#### 3) Qual o range de datas dos dados presentes na tabela *trips*?

```python
con.execute("""

    SELECT
        MIN(strptime(start_time, '%m/%d/%Y %H:%M:%S')) AS data_min,
        MAX(strptime(start_time, '%m/%d/%Y %H:%M:%S')) AS data_max
    FROM
        trips

""").fetchdf()
```

#### 4) Qual a média de duração das viagens de bike?

```python
con.execute("""

    SELECT
        AVG(duration_minutes) AS média
    FROM
        trips

""").fetchdf()
```

#### 5.1) Conte o total de viagens por estação de partida e ordene a contagem de forma descendente. Qual **station_id** teve mais partidadas de viagens?

```python
con.execute("""

    SELECT
        start_station_id,
        COUNT(DISTINCT trip_id) AS qtd
    FROM
        trips
    GROUP BY
        start_station_id
    ORDER BY
        qtd DESC

""").fetchdf()
```

#### 5.2) Qual o nome da estação que teve o maior número de partidas de viagens?
Dica: modifique a query anterior.



```python
con.execute("""

    SELECT
        s.name,
        COUNT(DISTINCT t.trip_id) AS qtd
    FROM
        trips AS t
    LEFT JOIN stations AS s
    ON t.start_station_id = s.station_id
    GROUP BY
        s.name
    ORDER BY
        qtd DESC

""").fetchdf()
```

#### 6.1) Conte o total de viagens por estação de chegada

```python
con.execute("""

    SELECT
        end_station_id,
        COUNT(DISTINCT trip_id) AS qtd
    FROM
        trips
    GROUP BY
        end_station_id
    ORDER BY
        qtd DESC

""").fetchdf()
```

#### 6.2) Qual o nome da estação que teve o maior número de chegadas de viagens?

```python
con.execute("""

    SELECT
        s.name,
        COUNT(DISTINCT t.trip_id) AS qtd
    FROM
        trips AS t
    LEFT JOIN stations AS s
    ON t.end_station_id = s.station_id
    GROUP BY
        s.name
    ORDER BY
        qtd DESC

""").fetchdf()
```

#### 7) Quais são as 5 maiores estações em área? Exiba o id, nome, número de docks e área total. Filtre somente areas não nulas e ordene de forma descendente

```python
con.execute("""

    SELECT
        station_id,
        name,
        number_of_docks,
        footprint_length * footprint_width AS area_total
    FROM 
        stations
    WHERE
        footprint_length IS NOT NULL AND footprint_width IS NOT NULL
    ORDER BY
        area_total DESC

""").fetchdf()
```

#### 8) Quais são os tipos de subscribers mais populares? Conte o total por tipo e ordene de forma descendente.

```python
con.execute("""

    SELECT
        subscriber_type,
        COUNT(*) AS qtd
    FROM trips
    GROUP BY
        subscriber_type
    ORDER BY
        qtd DESC

""").fetchdf()
```

#### 9) Quantas estações desativadas e ativadas existem na tabela?

```python
con.execute("""

    SELECT
        status,
        COUNT(*) AS qtd
    FROM
        stations
    GROUP BY
        status
    ORDER BY
        qtd DESC

""").fetchdf()
```
